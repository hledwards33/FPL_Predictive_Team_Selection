Note that team_x has around 50% blank entries. We deal with this below standardising the players team. All players will have only one team and it will be set as the latest team stored for that player
<br><br>
To do this we need to sort the dataset by date (kickoff_time).

```python
df['kickoff_time'][0][-9:-1] #slice out time
```

```python
df['kickoff_time'][0][:-10] #slice out date
```

```python
df['match_date'] = pd.to_datetime(df['kickoff_time'].apply(lambda x: x[:-10]) + " " + df['kickoff_time'].apply(lambda x: x[-9:-1]))
df.drop('kickoff_time', inplace=True, axis=1)
df['match_date'].head()
```

```python
#sort values by match date
df.sort_values(by=['match_date'], ascending=False, inplace=True)

#create dataset of only name instance
player_team = df.drop_duplicates(subset=['name'])[['name', 'team_x']]
player_team.reset_index(inplace=True, drop=True)
len(player_team.index)
```

```python
player_team['team_x'].isna().sum() #check for null team values
```

```python
player_team.head()
```

Merge datasets with most recent player teams

```python
df = pd.merge(df, player_team, how='left', on=['name']).rename({'team_x_y':'team'})
df.drop('team_x_x', axis=1, inplace=True)
df.head()
```
