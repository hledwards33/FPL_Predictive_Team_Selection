Note that team_x has around 50% blank entries. We deal with this below standardising the players team. All players will have only one team and it will be set as the latest team stored for that player
<br><br>
To do this we need to sort the dataset by date (kickoff_time).

```python
df['kickoff_time'][0][-9:-1] #slice out time
```

```python
df['kickoff_time'][0][:-10] #slice out date
```

```python
df['match_date'] = pd.to_datetime(df['kickoff_time'].apply(lambda x: x[:-10]) + " " + df['kickoff_time'].apply(lambda x: x[-9:-1]))
df.drop('kickoff_time', inplace=True, axis=1)
df['match_date'].head()
```

```python
#sort values by match date
df.sort_values(by=['match_date'], ascending=False, inplace=True)

#create dataset of only name instance
player_team = df.drop_duplicates(subset=['name'])[['name', 'team_x']]
player_team.reset_index(inplace=True, drop=True)
len(player_team.index)
```

```python
player_team['team_x'].isna().sum() #check for null team values
```

```python
player_team.head()
```

Merge datasets with most recent player teams

```python
df = pd.merge(df, player_team, how='left', on=['name']).rename(columns={'team_x_y':'team'})
df.drop('team_x_x', axis=1, inplace=True)
df.head()
```

```python
len(df.index) #check all rows are retained after merge
```

```python
#Check for missing values again
df.isna().sum()
```

Analysis the missing team_a_score and team_h_score

```python
df[(df['team_a_score'].isna()) | (df['team_h_score'].isna())]
```

Note the only missing values are for Arsenal vs Man City on the 11/03/20 at 7:30. This match does not exist. hence we can delete these rows.

```python
df.dropna(axis=0, inplace=True)
len(df.index) #check only 49 rows have been removed
```

The only column that may be useful to our is team and opposition team. So we will create dummy variables for team column. We will create a ranking system for opp_team_name.

```python
df['opp_team_name'].value_counts()
```

```python
df['season_x'].value_counts()
```

```python
data_path = os.getcwd() + "\Data\Raw_Data\EPL_Tables.csv"

prem_tables = pd.read_csv(data_path)
prem_tables.head()
```

```python
prem_tables['Season'].value_counts()
```

```python
#Score teams based on which quarter of the table they are in
def quartile_scoring(x):
    
    if x <=5:
        return 1
    elif x <=10:
        return 2
    elif x<=15:
        return 3
    else:
        return 4

prem_tables['opp_team_rating'] = prem_tables['Position'].apply(lambda x: quartile_scoring(x))
#drop Position column
prem_tables.drop('Position', axis=1, inplace=True)
prem_tables.head()
```

```python
#match key column names
df.rename(columns={'season_x': 'Season'}, inplace=True)
prem_tables.rename(columns={'Team': 'opp_team_name'}, inplace=True)

df = pd.merge(df, prem_tables, how='left', on=['Season', 'opp_team_name']).drop(['opp_team_name', 'opp_team'])
df.head()
```

Now we have added in Rankings for the opponend team we will now, create dummy variables for the player's team.

```python
dummies = pd.get_dummies(df['team'],drop_first=False)
#df = df.drop(['team'],axis=1) #do not drop team column as we will use this later on
df = pd.concat([df,dummies],axis=1)
df.head()
```
